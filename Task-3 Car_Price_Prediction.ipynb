```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("car data.csv")
```

```python
data
```

```python
#Getting summary information about the data
data.info()
```

```python
#shape of the data
data.shape
```

```python
#checking null values
data.isnull().sum()
```

```python
#Statistics about the data
data.describe()
```

```python
#Data Cleaning
data=data.drop(['Year','Selling_Price','Owner'],axis=1)
```

```python
data
```

```python
#Label Encoding
from sklearn.preprocessing import LabelEncoder,OneHotEncoder
lc=LabelEncoder()
```

```python
data['Car_Name']=lc.fit_transform(data['Car_Name'])
data['Fuel_Type']=lc.fit_transform(data['Fuel_Type'])
data['Selling_type']=lc.fit_transform(data['Selling_type'])
data['Transmission']=lc.fit_transform(data['Transmission'])
```

```python
data
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
```

```python
scaled_data=scaler.fit_transform(data)
```

```python
scaled_data=pd.DataFrame(scaled_data)
```

```python
scaled_data.columns=data.columns
```

```python
scaled_data
```

```python
x=scaled_data[['Car_Name','Driven_kms','Fuel_Type','Selling_type','Transmission']]
y=scaled_data['Present_Price']
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=3)
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
```

```python
model.fit(x_train,y_train)
```

```python
lr_pred=model.predict(x_test)
```

```python
lr_pred
```

```python
from sklearn.metrics import mean_absolute_error,r2_score,mean_squared_error
```

```python
#r2_score
print("R2_score is :",r2_score(lr_pred,y_test))
```

```python
import numpy as np
#Root mean squared error
print("Root mean squared error is :",np.sqrt(mean_squared_error(lr_pred,y_test)))
```
